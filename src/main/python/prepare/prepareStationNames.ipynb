# direction ba (Borgweg --> Arenen)

```python
import pandas as pd

# Load the CSV file
file_path = '..\..\..\..\scenarios\input\stationen_raw.csv'
data = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the dataframe to understand its structure
data.head(3)
```

```python
# Define the function to fill in the missing values based on the pattern
def fill_missing_values(row):
    if pd.isna(row['NodeNamen']):
        row['NodeNamen'] = row['Station'] + 'Node'
    if pd.isna(row['NodeID']):
        row['NodeID'] = row['Station'] + 'Node'
    if pd.isna(row['LinkName']):
        row['LinkName'] = 'BA-' +  row['Station'] + 'Link'
    if pd.isna(row['LinkID']):
        row['LinkID'] = 'BA-' +  row['Station'] + 'Link'
    if pd.isna(row['StopFacility']):
        row['StopFacility'] = 'BA-' + row['Station'] + 'StopFacility'
    if pd.isna(row['StopName']):
        row['StopName'] = 'BA-' + row['Station'] + 'Stop'
    return row

# Apply the function to fill missing values
data = data.apply(fill_missing_values, axis=1)

# Display the modified dataframe
data
```

```python
# Save the modified dataframe to a new CSV file
output_path = '..\..\..\..\scenarios\input\stationen_ba.csv'
data.to_csv(output_path, index=False, sep=',')
```

# direction ab (Arenen --> Borgweg)

```python
# Load the CSV file
file_path = '..\..\..\..\scenarios\input\stationen_raw.csv'
data_rev = pd.read_csv(file_path, delimiter=';')

# reihenfolge umdrehen
data_rev = data_rev.iloc[::-1]
data_rev = data_rev.reset_index(drop=True)
data_rev['departureOffset'] = data['departureOffset']
data_rev['arrivalOffset'] = data['arrivalOffset']

# Display the first few rows
data_rev.head(3)
```
