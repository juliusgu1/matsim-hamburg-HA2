```python
# MERGE VEHICLES
import xml.etree.ElementTree as ET

def merge_vehicle_files(existing_file, new_file):
    # Parse existing and new XML files
    existing_tree = ET.parse(existing_file)
    new_tree = ET.parse(new_file)

    # Find the root elements
    existing_root = existing_tree.getroot()
    new_root = new_tree.getroot()

    # Find the vehicleDefinitions element
    vehicle_definitions = existing_root.find('{http://www.matsim.org/files/dtd}vehicleDefinitions')

    # Check if U5_VehicleType already exists
    u5_vehicle_type_exists = False
    for vehicle_type in vehicle_definitions.findall('{http://www.matsim.org/files/dtd}vehicleType'):
        if vehicle_type.get('id') == 'U5_Vehicle':
            u5_vehicle_type_exists = True
            break

    # If U5_VehicleType doesn't exist, add it from new_file at the beginning
    if not u5_vehicle_type_exists:
        # Find U5_VehicleType in new_file
        new_vehicle_type = new_root.find('{http://www.matsim.org/files/dtd}vehicleType[@id="U5_Vehicle"]')
        
        if new_vehicle_type is not None:
            # Insert U5_VehicleType at the beginning of vehicleDefinitions
            vehicle_definitions.insert(0, new_vehicle_type)

    # Append new vehicles from new_file to existing_file
    for vehicle in new_root.findall('{http://www.matsim.org/files/dtd}vehicle'):
        existing_root.append(vehicle)

    return existing_tree

def save_merged_file(tree, output_file):
    # Register namespace to avoid default_namespace option error
    ET.register_namespace('', "http://www.matsim.org/files/dtd")
    # XML-Dokument formatieren
    format_xml(tree.getroot())
    # Write the merged XML tree to file
    tree.write(output_file, encoding='utf-8', xml_declaration=True)


# Funktion zum Formatieren des XML-Dokuments
def format_xml(element, level=0):
    indent = "    "
    if len(element):
        if not element.text or not element.text.strip():
            element.text = "\n" + indent * (level + 1)
        if not element.tail or not element.tail.strip():
            element.tail = "\n" + indent * level
        for elem in element:
            format_xml(elem, level + 1)
        if not element.tail or not element.tail.strip():
            element.tail = "\n" + indent * level
    else:
        if level and (not element.tail or not element.tail.strip()):
            element.tail = "\n" + indent * level

# Beispielaufruf
merged_vehicles_tree = merge_vehicle_files(existing_vehicles_file, new_vehicles_file)
save_merged_file(merged_vehicles_tree, merged_vehicles_file)
```
